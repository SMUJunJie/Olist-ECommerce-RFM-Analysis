```python
customers = pd.read_csv("olist_customers_dataset.csv")
orders = pd.read_csv("olist_orders_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
```

```python
sql_base_data = """

WITH base_data AS (
    SELECT 
        c.customer_unique_id,
        o.order_id,
        o.order_purchase_timestamp,
        p.payment_value
    FROM customers c
    LEFT JOIN orders o
        ON c.customer_id = o.customer_id
    LEFT JOIN payments p
        ON o.order_id = p.order_id
    WHERE o.order_purchase_timestamp <= '2018-09-01'
)

SELECT
    customer_unique_id,
    DATEDIFF('day', CAST(MAX(order_purchase_timestamp) AS DATE), DATE '2018-09-01') AS recency_days, 
    COUNT(DISTINCT order_id) AS amount,
    SUM(payment_value) AS total_value
FROM base_data
GROUP BY customer_unique_id
ORDER BY total_value DESC


"""
```

```python
duckdb.sql(sql_base_data).show()
```

```python
sql_data = """

WITH base AS (
    SELECT 
        c.customer_unique_id,
        DATE_DIFF('day', CAST(MAX(order_purchase_timestamp) AS DATE), DATE '2018-09-01') AS recency_days,
        COUNT(DISTINCT o.order_id) AS frequency,
        SUM(payment_value) AS monetary
    FROM customers c
    JOIN orders o ON c.customer_id = o.customer_id
    JOIN payments p ON o.order_id = p.order_id
    WHERE o.order_purchase_timestamp <= '2018-09-01'
    GROUP BY c.customer_unique_id
),

scores AS (
SELECT 
    *,
    NTILE(5) OVER(ORDER BY recency_days DESC) AS r_score,
    NTILE(5) OVER(ORDER BY frequency DESC) AS f_score,
    NTILE(5) OVER(ORDER BY monetary DESC) AS m_score
FROM base
)

SELECT 
    customer_unique_id,
    recency_days,
    frequency,
    monetary,
    CONCAT(r_score, r_score, r_score) AS score
    FROM scores
    ORDER BY monetary DESC
"""
```

```python
duckdb.sql(sql_data).show()
```
